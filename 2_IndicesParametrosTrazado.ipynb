```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv('/content/1695762808952-Population.csv',encoding='UTF-8-sig',sep=',')
population.head()
```

```python
countries = ['Mexico', 'United States', 'Canada']
northAmerica = population.loc[population['Country Name'].isin(countries)]
northAmerica = northAmerica.drop(['Country Code'],axis=1)
northAmerica= northAmerica.set_index('Country Name')
northAmerica.head()
```

```python
populationMelt = pd.read_csv('/content/1695762808952-Population.csv',encoding='UTF-8-sig',sep=',')
columns=['Country Name','Country Code','2019','2018','2017']
countries=['Canada','Germany','Japan','Mexico','Nigeria','United States']
populationMelt = populationMelt.loc[populationMelt['Country Name'].isin(countries)][columns]
populationMelt.set_index('Country Code',inplace=True)
populationMelt.head()
```

```python
populationMelt.reset_index(inplace=True)
populationMelt=populationMelt.drop(['Country Name'],axis=1)
df = populationMelt.melt(id_vars='Country Code',var_name='Year',value_name='Population')
df.head()
```

```python
## graficar con seaborn

sns.barplot(x='Country Code',y='Population',hue='Year',data=df)
```

```python
populationDesvStd = pd.read_csv('/content/1695762808952-Population.csv',encoding='UTF-8-sig',sep=',')
columns=['Country Name','Country Code','2019','2018','2017']
countries=['Canada','Germany','Japan','Mexico','Nigeria','United States']
populationDesvStd = populationDesvStd.loc[populationDesvStd['Country Name'].isin(countries)][columns]
populationDesvStd.set_index('Country Code',inplace=True)

populationDesvStd.head()
populationDesvStd.reset_index(inplace=True)
populationDesvStd=populationDesvStd.drop(['Country Name'],axis=1)
populationDesvStdGroup1 = populationDesvStd.melt(id_vars='Country Code',var_name='Year',value_name='Population')
populationDesvStdGroup = populationDesvStd.melt(id_vars='Country Code',var_name='Year',value_name='Population')

populationDesvStdGroup =populationDesvStdGroup.groupby('Country Code').std()
populationDesvStdGroup.head()
```

```python
## con matplotlib
plt.bar(populationDesvStdGroup.index,populationDesvStdGroup['Population'])
```

```python
## con pandas
populationDesvStdGroup.plot.bar()
```

```python
## con seaborn

sns.barplot(x='Country Code',y='Population',data=populationDesvStdGroup)
```

```python
## con seaborn obteniendo la media directo en el grafico

sns.barplot(x='Country Code',y='Population',data=populationDesvStdGroup1,estimator=np.std,ci=None)
```
