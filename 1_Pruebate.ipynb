```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv('/content/1695677925032-Fertility.csv',encoding='UTF-8-sig',sep=',')
population.head()
```

```python
mexico = population.loc[population['Country Name']=='Mexico']
mexico = mexico.drop(['Country Code'],axis=1)
mexico = mexico.set_index('Country Name')
mexico = mexico.T
mexico.columns = ['Fertility']
mexico.index.name = 'Year'
mexico.head()

```

```python
mexico.plot(title = 'Fertility rate in Mexico',style = '--')
plt.grid(True, color='blue')
```

```python
MexicoCanada = population.loc[(population['Country Name']=='Mexico')|(population['Country Name']=='Canada')]
MexicoCanada = MexicoCanada.drop(['Country Code'],axis=1)
MexicoCanada = MexicoCanada.set_index('Country Name')
MexicoCanada = MexicoCanada.T
#MexicoCanada.columns = ['Fertility']
#MexicoCanada.index.name = 'Year'
MexicoCanada.head()
```

```python
MexicoCanada.plot(title = 'Fertility rate',style = '--')
plt.grid(True, color='gray')
```
