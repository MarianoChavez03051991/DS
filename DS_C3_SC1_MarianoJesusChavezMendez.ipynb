```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Metadata = pd.read_csv('/content/0.9ep3krj7s1o0.ona2jvufk4mMetadata.csv',encoding='UTF-8-sig',sep=',')
Metadata
```

```python
Happiness_report = pd.read_csv('/content/0.rw6r6mdtjng0.9sxn8dzeqsbHappiness_report.csv',encoding='UTF-8-sig',sep=',')
Happiness_report.head()
```

```python
Happiness_World = Happiness_report
Happiness_World = Happiness_World.set_index('Country or region')
Happiness_World.head()
```

```python

Happiness_Mexico = Happiness_report[Happiness_report['Country or region']=='Mexico']
Happiness_Mexico = Happiness_Mexico.set_index('Country or region')
Happiness_Mexico.head()
```

```python
value = Happiness_Mexico.loc['Mexico','Score']
value
```

```python
sns.set_style('white')
props_bbox = {'boxstyle': 'circle','facecolor': 'red','edgecolor':'black','alpha': 0.5}

sns.histplot(x='Score',data=Happiness_World,legend=True)
plt.title('World Happiness Report', fontsize=14, weight='bold')
plt.ylabel('Frequency')

value = Happiness_Mexico.loc['Mexico','Score']
plt.text(value,1,'Mex',color='white',bbox=props_bbox)

plt.show()
```

```python
mex_value = Happiness_Mexico.loc[Happiness_Mexico.index=='Mexico','Healthy life expectancy'].to_list()[0]
mex_value
```

```python
sns.set_style('white')

props_bbox = {'boxstyle': 'circle','facecolor': 'red','edgecolor':'black','alpha': 0.5}
props_arrow = {'arrowstyle':'->','color':'black'}

sns.catplot(kind='box',y='Healthy life expectancy',color='white',data=Happiness_report,orient='h',capprops=dict(color='black')
,flierprops=dict(marker='o', markerfacecolor='white', markersize=10, markeredgecolor='black'))

mex_value = Happiness_Mexico.loc[Happiness_Mexico.index=='Mexico','Healthy life expectancy'].to_list()[0]

plt.title('World Happiness Report', fontsize=14, weight='bold')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('')

plt.annotate('MEX', xy=(0,mex_value),xytext=(0.5,mex_value+0.1),color='white',arrowprops=props_arrow,bbox=props_bbox)

plt.show()
```

```python
Happiness_MexicoPie = Happiness_Mexico
Happiness_MexicoPie = Happiness_MexicoPie.T
Happiness_MexicoPie = Happiness_MexicoPie.loc[Happiness_MexicoPie.index!='Overall rank']
Happiness_MexicoPie = Happiness_MexicoPie.loc[Happiness_MexicoPie.index!='Score']
Happiness_MexicoPie.columns = ['Factors']
Happiness_MexicoPie.index.name = 'Values'
Happiness_MexicoPie.head()

Happiness_MexicoPie1 = Happiness_MexicoPie.T
Happiness_MexicoPie1.head()
```

```python
sns.set_style('white')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), gridspec_kw={'width_ratios': [1, 3]})
fig.suptitle('Contribution in Mexico happines score',fontsize=16)


table = ax1.table(cellText=Happiness_MexicoPie1.T.values,rowLabels=Happiness_MexicoPie1.columns,colLabels=Happiness_MexicoPie1.index,loc='lower left')
table.auto_set_font_size(False)
table.scale(1.2, 1.2)
ax1.axis('off')


Happiness_MexicoPie['Factors'].plot(kind='pie',autopct='%1.1f%%',ax=ax2)
plt.xlabel('')
plt.ylabel('')




```

```python
Happiness_WorldTopTwo = Happiness_World.sort_values(by='Overall rank',ascending=True)
Happiness_WorldTopTwo = Happiness_WorldTopTwo.iloc[:2]
Happiness_WorldTopMexico = Happiness_World.loc[Happiness_World.index=='Mexico']
countries = ('Canada','United States')
Happiness_WorldTopAddTwo = Happiness_World.loc[Happiness_World.index.isin(countries)]
```

```python
Happiness_WorldTop = pd.concat([Happiness_WorldTopTwo,Happiness_WorldTopMexico,Happiness_WorldTopAddTwo])
Happiness_WorldTop = Happiness_WorldTop.reset_index()
Happiness_WorldTop
```

```python
Happiness_WorldTop = pd.concat([Happiness_WorldTopTwo,Happiness_WorldTopMexico,Happiness_WorldTopAddTwo])
Happiness_WorldTop = Happiness_WorldTop.reset_index()
Happiness_WorldTop = Happiness_WorldTop.melt(id_vars='Country or region',var_name='Factors',value_name='Values')
Happiness_WorldTop = Happiness_WorldTop.loc[Happiness_WorldTop['Factors']!='Overall rank']
Happiness_WorldTop = Happiness_WorldTop.loc[Happiness_WorldTop['Factors']!='Score']
Happiness_WorldTop.head()
```

```python
sns.set_style('white')

sns.barplot(x='Factors', y='Values', hue='Country or region', data=Happiness_WorldTop)
plt.xticks(rotation=90)
plt.title('World Happiness Report', fontsize=14, weight='bold')
plt.xlabel('')
plt.ylabel('')
```

```python
Happiness_WorldxMetadata = pd.merge(left=Happiness_World,right=Metadata,left_index=True,right_index=True)
Happiness_WorldxMetadata.reset_index(inplace=True)
Happiness_WorldxMetadata.head()
```

```python
colums=['Region','Score']
Happiness_WorldxMetadata = Happiness_WorldxMetadata[colums]
Happiness_WorldxMetadata = Happiness_WorldxMetadata.groupby('Region').mean('Score')
Happiness_WorldxMetadata.head()
```
