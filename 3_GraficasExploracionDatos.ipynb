```python
newMergeData2 = pd.merge(left=newMergeData,right=metaData,left_on='Country Code',right_on='Country Code')
newMergeData2.head()
```

```python
## con pandas y matplotlib para separa por grupos es necesario iterar cada grupo sobre el dataframe

groups = newMergeData2.groupby('Region')

for name,group in groups:
  plt.scatter(x=group['2019_GNI'],y=group['2019_lifeExpentancy'],label=name)
  plt.legend()
  plt.xscale('log')
```

```python
## con seaborn si es plsible a traves de la propiedad hue

splot= sns.scatterplot(x='2019_GNI',y='2019_lifeExpentancy',data=newMergeData2,hue='IncomeGroup')
splot.set(xscale='log')
```

```python
## con seaborn si es plsible separarlo a traves  de la propiedad col

splot= sns.relplot(kind='scatter',x='2019_GNI',y='2019_lifeExpentancy',data=newMergeData2,col='IncomeGroup')
splot.set(xscale='log')
```
