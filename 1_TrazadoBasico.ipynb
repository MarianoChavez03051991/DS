```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv('/content/1695762808952-Population.csv',encoding='UTF-8-sig',sep=',')
population.head()
```

```python
mexico = population.loc[population['Country Name']=='Mexico','2014':'2019']
mexico.head()
```

```python
mexico = mexico.T
mexico.head()
```

```python
##matplotlib
plt.plot(mexico.index,mexico.values)
```

```python
##pandas
mexico.plot()
```

```python
##seaborn
sns.lineplot(x=mexico.index,y=mexico[152])
```

```python
##personalizaciÃ³n de la grafica en matplotlib
plt.plot(mexico.index,mexico.values,color='r',linestyle='--',marker='o')
```

```python
##personalzaciÃ³n de la grafica en pandas
mexico.plot(title='Population of Mexico',
legend=False,xlabel='Year',ylabel='Population', rot=45,fontsize=16)
```
